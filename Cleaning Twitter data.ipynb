```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import math
import os
import plotly.graph_objects as go
from datetime import date
from datetime import datetimes
import warnings
import requests
from textblob import TextBlob
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"/Users/lindseytam/Desktop/team-shows/geoTwitter18-01-01/geoTwitter18-01-01_12.csv") 
del df['Unnamed: 0']
df.head()
```

```python
words = ["bachelor", "JanuARIE", "bachelornation", "ariejr"]

csv_dir = "/Users/lindseytam/Desktop/team-shows/geoTwitter18-01-01/"
data = pd.DataFrame(columns=['text', 'date', 'city', 'state'])


for filename in os.listdir(csv_dir):
    csv_path = os.path.join(csv_dir, filename)
    try:
        df = pd.read_csv(csv_path, error_bad_lines=False, encoding='utf-8', engine='python')
    except Exception:
        print(csv_path)
        os.remove(csv_path)
        continue

    for i in range(len(df)):
        for word in words:
            if word in df["text"][i]:
                data = data.append(df.iloc[i])
        
data.head()
```

```python
data.index = np.arange(0, len(data) )
polarity = []

for i in range(len(data)):
    polarity.append(TextBlob(data["text"][i]).sentiment.polarity)
data["polarity"] = polarity
del data['Unnamed: 0']
data.head()
```
